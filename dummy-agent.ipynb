```python
from agents import Agent, function_tool,Runner, ItemHelpers


@function_tool
def get_weather(city: str) -> str:
    """현재 도시의 날씨 정보를 조회합니다."""
    # 실제 구현 시 OpenWeatherMap 등 API 연동 가능
    return f"{city}의 현재 날씨: 맑음, 기온 22°C, 습도 45%"


weather_agent = Agent(
    name="Weather Agent",
    instructions="날씨 관련 질문에 답변합니다. 도시 이름을 받으면 get_weather 도구로 날씨를 조회하고, 사용자에게 친절하게 알려주세요.",
    tools=[get_weather],
)

result = Runner.run_streamed(weather_agent, "서울 날씨 어때?")

async for event in result.stream_events() :
    #tool을 사용하거나 response를 주려고 할때 타입
    #agent에게 무슨일이 일어나고 있는지 더 자세하게 제공한다
    if event.type == 'raw_response_event':
        continue
    #agent가 변경되었을때
    elif event.type == "agent_updated_stream_event":
        print("Agent updated to : ", event.new_agent.name)
    #작업 단위가 끝났을때 이벤트
    elif event.type == "run_item_stream_event":
        #print(event.item.type)
        if event.item.type == "tool_call_item":
            print("Tool call : ", event.item.raw_item.to_dict())
        elif event.item.type == "tool_call_output_item":
            print("Tool result : ", event.item.output)
        elif event.item.type == "message_output_item":
            print(ItemHelpers.text_message_output(event.item))
    print("="*20)
#print(result.final_output)
```

```python
from agents import Agent, function_tool,Runner, ItemHelpers


@function_tool
def get_weather(city: str) -> str:
    """현재 도시의 날씨 정보를 조회합니다."""
    # 실제 구현 시 OpenWeatherMap 등 API 연동 가능
    return f"{city}의 현재 날씨: 맑음, 기온 22°C, 습도 45%"


weather_agent = Agent(
    name="Weather Agent",
    instructions="날씨 관련 질문에 답변합니다. 도시 이름을 받으면 get_weather 도구로 날씨를 조회하고, 사용자에게 친절하게 알려주세요.",
    tools=[get_weather],
)

result = Runner.run_streamed(weather_agent, "서울 날씨 어때?")

async for event in result.stream_events() :
    #tool을 사용하거나 response를 주려고 할때 타입
    #agent에게 무슨일이 일어나고 있는지 더 자세하게 제공한다
    if event.type == 'raw_response_event':
        #실제로 쓰여지는 문자를 보여줌
        if event.data.type == "response.output_text.delta":
            print(event.data.delta)
    print("="*20)
#print(result.final_output)
```
